```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 14,7
plt.style.use('seaborn-whitegrid')


address =  "/content/drive/MyDrive/Colab Notebooks/Data/mtcars.csv"
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
X = cars[['cyl','vs','am','gear']]
sb.pairplot(X)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coef, p_value = scipy.stats.spearmanr(cyl,vs)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coef))
```

```python
spearmanr_coef, p_value = scipy.stats.spearmanr(cyl,am)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coef))
```

```python
spearmanr_coef, p_value = scipy.stats.spearmanr(cyl,gear)

print('Spearman Rank Correlation Coefficient %0.3f' % (spearmanr_coef))
```

```python
table = pd.crosstab(cyl,am)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```
