```python
%cd "/content/drive/MyDrive/Colab Notebooks/freeCodeCamp-ML"
from __future__ import absolute_import, division, print_function, unicode_literals
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output
from six.moves import urllib

import tensorflow.compat.v2.feature_column as fc
import tensorflow as tf
```

```python
x_train = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/train.csv')
x_test = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/eval.csv')

# pop() columnu remove eder ve döner
y_train = x_train.pop('survived')
y_test = x_test.pop('survived')
```

```python
CATEGORICAL_COLUMNS = ['sex','n_siblings_spouses','parch','class','deck','embark_town','alone']
NUMERIC_COLUMNS = ['age','fare']
```

```python
X = []
for feature in CATEGORICAL_COLUMNS:
  vocabulary = x_train[feature].unique()
  X.append(tf.feature_column.categorical_column_with_vocabulary_list(feature,vocabulary))

print(X)
```

```python
def make_inp_fn(data,label,epochs=10, shuffle=True, batch_size=32):
  def input_function():
    ds = tf.data.Dataset.from_tensor_slices((dict(data),label))
    if shuffle:
      ds = ds.shuffle(1000)
    ds = ds.batch(batch_size).repeat(epochs)
    return ds
  return input_function
```

```python
train_fn = make_inp_fn(x_train,y_train)
test_fn = make_inp_fn(x_test, y_test,epochs=1,shuffle=False)
```

```python
linear_est = tf.estimator.LinearClassifier(feature_columns=X)
linear_est.train(train_fn)
result = linear_est.evaluate(test_fn)

clear_output()
print(result['accuracy'])
```
