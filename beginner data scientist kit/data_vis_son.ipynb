```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from pylab import rcParams
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

```python
df = pd.read_csv("Superstore-Sales.csv", index_col="Order Date",encoding='cp1252',parse_dates=True)
df.head()
```

```python
df['Order Quantity'].plot()
```

```python
df2 = df.sample(n=100, random_state=25, axis=0)
plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Supersstore Sales')

df2['Order Quantity'].plot()
```

```python
import seaborn as sb
sb.set_style('whitegrid')

cars = pd.read_csv("mtcars.csv")
cars.columns = ['car_names','mpg','cyl', 'disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.index = cars.car_names
mpg = cars['mpg']
mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sb.distplot(mpg)
```

```python
cars.plot(kind='scatter',x='hp',y='mpg',c=['darkgray'],s=150)
```

```python
sb.regplot(x='hp',y='mpg',data=cars,scatter=True)
```

```python
sb.pairplot(cars)
```

```python
temp = cars[['mpg','disp','hp','wt']]
sb.pairplot(temp)
plt.show()
```

```python
cars.boxplot(column='mpg',by='am')
cars.boxplot(column='wt',by='am')
```

```python
sb.boxplot(x='am',y='mpg',data=cars, palette='hls')
```

```python
plt.plot(cars['am'],cars['mpg'],marker="s",ls="--",mew=10)
```
