```python
%cd "/content/drive/MyDrive/Colab Notebooks/freeCodeCamp-ML"
from __future__ import absolute_import, division, print_function, unicode_literals

import tensorflow as tf
import pandas as pd
```

```python
COLUMNS = ['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']
SPECIES = ['Setosa','Versicolor','Virginica']
```

```python
train_path = tf.keras.utils.get_file("iris_training.csv","https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv")
test_path = tf.keras.utils.get_file("iris_test.csv","https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv")

X_train = pd.read_csv(train_path,names=COLUMNS, header=0)
X_test = pd.read_csv(test_path,names=COLUMNS,header=0)
```

```python
y_train = X_train.pop('Species')
y_test = X_test.pop('Species')
X_train.head()
```

```python
def input_function(data,labels,training=True,batch_size=256):
    ds = tf.data.Dataset.from_tensor_slices((dict(data),labels))
    if training:
      ds = ds.shuffle(1000).repeat()
   
    return ds.batch(batch_size)
```

```python
X = []
for key in X_train.keys():
  X.append(tf.feature_column.numeric_column(key=key))

print(X)
```

```python
from tensorflow._api.v2 import feature_column
clf = tf.estimator.DNNClassifier(feature_columns=X, hidden_units=[30,10], n_classes=3)

clf.train(input_fn=lambda: input_function(X_train, y_train, training=True), steps=5000)
```

```python
eval_result = clf.evaluate(input_fn= lambda: input_function(X_test,y_test,training=False))
print("\nTest set accuracy: {accuracy:0.3f}\n".format(**eval_result))
```

```python
def input_fn(features, batch_size=256):
  return tf.data.Dataset.from_tensor_slices(dict(features)).batch(batch_size)

features = ['SepalLength','SepalWidth','PetalLength','PetalWidth']
predict = {}

print("Please type numeric values as promted.")
for feature in features:
  val = input(feature+": ")
  try:
    predict[feature] = [float(val)]
  except ValueError:
    print("value error try again")
    break


  

predictions = clf.predict(input_fn = lambda: input_fn(predict))
for pred_dict in predictions:
  class_id = pred_dict['class_ids'][0]
  probability = pred_dict['probabilities'][class_id]

  print('Prediction is "{}" ({:.1f}%)'.format(SPECIES[class_id], 100*probability))
```
