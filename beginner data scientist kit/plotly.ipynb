```python
cars_subset = cars[['mpg','disp','hp']]
from sklearn.preprocessing import StandardScaler
cars_data_std = StandardScaler().fit_transform(cars_subset)

cars_select = pd.DataFrame(cars_data_std)
cars_select.columns = ['mpg','disp','hp']

cars_select.iplot(kind='histogram',filename='multiple-hist-charts')
```

```python
cars_select.iplot(kind='histogram',filename='multiple-hist-charts',subplots=True)
```

```python
cars_select.iplot(kind='histogram',filename='multiple-hist-charts',subplots=True,shape=(3,1))
```

```python
cars_select.iplot(kind='histogram',filename='multiple-hist-charts',subplots=True,shape=(1,3))
```

```python
cars_select.iplot(kind='box',filename='multiple-box-charts',subplots=True,shape=(1,3))
```
