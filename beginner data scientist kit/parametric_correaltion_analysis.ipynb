```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
plt.style.use('seaborn-whitegrid')
```

```python
address =  "/content/drive/MyDrive/Colab Notebooks/Data/mtcars.csv"
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
sb.pairplot(cars)
```

```python
X = cars[['mpg','hp','qsec','wt']]
sb.pairplot(X)
```

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearson_coef, p_value = scipy.stats.pearsonr(mpg,hp)
print('PearsonR Correliaton Coeffiient %0.3f' % (pearson_coef))
```

```python
pearson_coef, p_value = scipy.stats.pearsonr(mpg,qsec)
print('PearsonR Correliaton Coeffiient %0.3f' % (pearson_coef))
```

```python
pearson_coef, p_value = scipy.stats.pearsonr(mpg,wt)
print('PearsonR Correliaton Coeffiient %0.3f' % (pearson_coef))
```
